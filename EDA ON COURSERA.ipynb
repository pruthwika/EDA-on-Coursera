```python
df = df.dropna()
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df['course_difficulty'].unique()
```

```python
difficulty_mapping = {'Beginner': 1, 'Intermediate': 2, 'Mixed': 3, 'Advanced': 4}
df['course_difficulty_numeric'] = df['course_difficulty'].map(difficulty_mapping)
df
```

```python
df['course_Certificate_type'].unique()
```

```python
certificate_mapping = {
    'Course': 1,
    'Guided Project': 2,
    'Professional Certificate': 3,
    'Specialization': 4
}

df['course_certificate_numeric'] = df['course_Certificate_type'].map(certificate_mapping)
df.head()
```

```python
# Encoding categorical variables
cat_values=[]
num_values=[]
for columns in df.columns:
    if df[columns].nunique()<=10:
        cat_values.append(columns)
    else:
        num_values.append(columns)
        
```
